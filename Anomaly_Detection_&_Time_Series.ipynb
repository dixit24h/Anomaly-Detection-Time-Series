**1.** What is Anomaly Detection? Explain its types (point contextual, and collective anomalies) with examples.

- Anomaly detection, also known as outlier detection, is the process of identifying rare items, events, or observations that deviate significantly from the majority of the data. These 'anomalies' often indicate critical incidents, such as a technical glitch, a potential fraud, or a new trend.

- Types of Anomalies:

1.  **Point Anomalies (Global Outliers):**
    *   An individual data point is anomalous with respect to the rest of the data.
    *   **Example:** A credit card transaction of $10,000 in a dataset where all other transactions are under $100.

2.  **Contextual Anomalies (Conditional Outliers):**
    *   A data point is anomalous in a specific context but not necessarily otherwise.
    *   **Example:** A temperature reading of 30°C in winter is anomalous, but the same temperature in summer is normal.

3.  **Collective Anomalies:**
    *   A collection of related data points is anomalous with respect to the entire dataset, even if individual data points within the collection are not anomalous.
    *   **Example:** A sequence of repetitive, small network requests from a single IP address that, individually, might not trigger an alert, but together indicate a Denial-of-Service (DoS) attack.

**2.** Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.

- **Isolation Forest:**
    *   **Approach:** This algorithm isolates anomalies by randomly selecting a feature and then randomly selecting a split value between the maximum and minimum values of the selected feature. This process is repeated to build 'isolation trees'. Anomalies are points that require fewer splits to be isolated.
    *   **Use Cases:** Highly effective for high-dimensional datasets and large datasets. It's robust to irrelevant features and does not rely on distance metrics.

- **DBSCAN (Density-Based Spatial Clustering of Applications with Noise):**
    *   **Approach:** DBSCAN groups together points that are closely packed together (points with many nearby neighbors), marking as outliers those points that lie alone in low-density regions. It defines core points, border points, and noise points (anomalies).
    *   **Use Cases:** Excellent for detecting arbitrarily shaped clusters and identifying outliers in spatial data. It doesn't require the number of clusters to be specified beforehand.

- **Local Outlier Factor (LOF):**
    *   **Approach:** LOF measures the local deviation of density of a given data point with respect to its neighbors. It considers as outliers those objects that have a substantially lower density than their neighbors. The 'local' aspect means it can detect anomalies in different density regions.
    *   **Use Cases:** Particularly good for detecting anomalies in datasets where the anomaly score depends on the local neighborhood rather than the global distribution. Suitable for varied density datasets.

**3.** What are the key components of a Time Series? Explain each with one example.

- A time series is a sequence of data points indexed (or listed or graphed) in time order. Most commonly, a time series is a sequence taken at successive equally spaced points in time. Key components often include:

1.  **Trend:**
    *   **Explanation:** The long-term increase or decrease in the data over time. It reflects the underlying direction of the series.
    *   **Example:** The general upward movement of global temperatures over several decades, indicating global warming.

2.  **Seasonality:**
    *   **Explanation:** Repetitive patterns or cycles that occur at fixed or known intervals (e.g., daily, weekly, monthly, yearly). These patterns are often influenced by calendar events or natural phenomena.
    *   **Example:** The spike in retail sales during the holiday season each year, or higher electricity consumption during summer afternoons.

3.  **Cyclicality:**
    *   **Explanation:** Patterns that represent rises and falls in the data but are not of fixed period, or of periods longer than a year. These are often related to economic cycles or business cycles.
    *   **Example:** Periods of economic expansion followed by recession, which can span several years and do not have a consistent duration.

4.  **Irregular (or Residual) Component:**
    *   **Explanation:** The random, unpredictable fluctuations in the data that cannot be explained by trend, seasonality, or cyclicality. This is often considered the 'noise' in the series.
    *   **Example:** A sudden, unexpected drop in stock prices due to unforeseen geopolitical events, which is not part of any regular pattern.

**4.** Define Stationary in time series. How can you test and transform a non-stationary series into a stationary one?

- **Stationary Time Series:**
    *   A time series is said to be stationary if its statistical properties (mean, variance, and autocorrelation) do not change over time. This means that the series has no trend, seasonal effects, or other time-dependent structures. Stationary series are easier to model and forecast because their behavior is consistent over time.

- **How to Test for Stationarity:**
    1.  **Visual Inspection:** Plot the time series and look for obvious trends, seasonality, or changing variance.
    2.  **Summary Statistics:** Divide the data into segments and compare the mean and variance across these segments. For a stationary series, these should be relatively constant.
    3.  **Statistical Tests:**
        *   **Augmented Dickey-Fuller (ADF) Test:** This is a popular statistical test that checks for the presence of a unit root in a time series. The null hypothesis is that the time series is non-stationary. If the p-value is below a chosen significance level (e.g., 0.05), you can reject the null hypothesis and conclude that the series is stationary.
        *   **Kwiatkowski-Phillips-Schmidt-Shin (KPSS) Test:** This test is the opposite of the ADF test. The null hypothesis is that the time series is stationary. If the p-value is below a chosen significance level, you reject the null hypothesis and conclude that the series is non-stationary.

- **How to Transform a Non-Stationary Series into a Stationary one:**
    1.  **Differencing:** This is the most common method. It involves computing the difference between consecutive observations. First-order differencing (`Xt - Xt-1`) can remove linear trends. Seasonal differencing (`Xt - Xt-s`, where 's' is the seasonal period) can remove seasonality.
    2.  **Transformation (e.g., Log Transform, Square Root Transform):** These can be used to stabilize the variance of a time series, especially when the variance increases with the mean. A common transformation is the natural logarithm (`log(Xt)`).
    3.  **Detrending:** If a clear trend is present, you can model the trend (e.g., using linear regression) and then subtract it from the original series to get the residuals.
    4.  **Seasonal Adjustment:** Similar to detrending, if strong seasonality is present, you can identify and remove the seasonal component using methods like seasonal decomposition of time series (STL, X-11, etc.).

**5.** Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in terms of structure and application.

- **AR (AutoRegressive) Model:**
    *   **Structure:** An AR model predicts future values based on a linear combination of its own past values. It's denoted as AR(p), where 'p' is the order of the model (number of past observations used).
    *   **Application:** Suitable for time series where the current value depends on past values, exhibiting memory or inertia.

- **MA (Moving Average) Model:**
    *   **Structure:** An MA model predicts future values based on a linear combination of past forecast errors (residuals). It's denoted as MA(q), where 'q' is the order of the model (number of past forecast errors used).
    *   **Application:** Suitable for time series where the current value depends on past random shocks or unexpected events.

- **ARIMA (AutoRegressive Integrated Moving Average) Model:**
    *   **Structure:** ARIMA(p, d, q) combines AR and MA models with an 'Integrated' component. 'd' represents the number of differencing steps needed to make the series stationary. This model handles non-stationary time series by differencing.
    *   **Application:** A versatile model for non-seasonal time series forecasting, especially when there are trends or non-stationarity that can be resolved by differencing.

- **SARIMA (Seasonal AutoRegressive Integrated Moving Average) Model:**
    *   **Structure:** SARIMA(p, d, q)(P, D, Q)s extends ARIMA to include seasonal components. (p, d, q) are non-seasonal orders, and (P, D, Q)s are seasonal orders, where 's' is the length of the seasonal period.
    *   **Application:** Ideal for time series with both non-seasonal and seasonal patterns, such as monthly sales data with annual seasonality.

- **SARIMAX (Seasonal AutoRegressive Integrated Moving Average with eXogenous regressors) Model:**
    *   **Structure:** SARIMAX is an extension of SARIMA that includes exogenous variables (other independent variables that can influence the time series). These external factors are included as linear predictors.
    *   **Application:** Used when forecasting a time series that is influenced by external factors not captured in the historical time series itself, like predicting sales with advertising spend as an exogenous variable.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose

# Load the AirPassengers dataset using get_rdataset
import statsmodels.api as sm
# Ensure the 'rpy2' package is installed if using get_rdataset without pre-installed R datasets
# !pip install rpy2 # uncomment if rpy2 is not installed and needed

data = sm.datasets.get_rdataset("AirPassengers", "datasets")
airline = data.data
airline.columns = ['Month', 'Passengers']

# The 'Month' column needs to be parsed as datetime and set as index
airline['Month'] = pd.to_datetime(airline['Month'])
airline = airline.set_index('Month')

# Plot the original series
plt.figure(figsize=(12, 6))
sns.lineplot(data=airline, x=airline.index, y='Passengers')
plt.title('Air Passengers Original Time Series')
plt.xlabel('Date')
plt.ylabel('Number of Passengers')
plt.grid(True)
plt.show()
```

### Decomposing the Time Series

```python
# Perform seasonal decomposition
# The model can be 'additive' or 'multiplicative'.
# For AirPassengers, which shows increasing variance with increasing trend, multiplicative is often more appropriate.
# freq parameter is important for decomposition; for monthly data, it's 12.

decomposition = seasonal_decompose(airline['Passengers'], model='multiplicative', period=12)

# Plot the decomposed components
fig = decomposition.plot()
fig.set_size_inches(10, 8)
plt.tight_layout()
plt.show()

# You can also access the components directly:
# trend = decomposition.trend
# seasonal = decomposition.seasonal
# residual = decomposition.resid
```

Next, we will select numerical features from the dataset and apply the Isolation Forest model to detect anomalies. We will use `fare_amount` and `passenger_count` as an example, but you can select other relevant numerical features.

```python
# Select features for anomaly detection
# For simplicity, let's use 'fare_amount' and 'passenger_count' for now.
# You might want to use more features or create new ones (e.g., trip distance).
features = ['fare_amount', 'passenger_count']
X = df_taxi[features].copy()

# Handle potential missing values (if any) - Isolation Forest can handle NaNs by default,
# but it's good practice to be aware or impute if necessary.
# For this example, we'll drop rows with NaNs in selected features for simplicity.
X = X.dropna()

# Initialize and train the Isolation Forest model
# contamination: The proportion of outliers in the data set. Used when fitting to define the threshold.
#                  If 'auto', the threshold is determined by the model.
# random_state for reproducibility
isolation_forest = IsolationForest(contamination='auto', random_state=42)
isolation_forest.fit(X)

# Calculate anomaly scores and predictions using the original features
anomaly_scores = isolation_forest.decision_function(X)
anomaly_predictions = isolation_forest.predict(X)

# Add anomaly scores and predictions as new columns to X
X['anomaly_score'] = anomaly_scores
X['anomaly'] = anomaly_predictions

# Separate inliers and outliers for plotting
inliers = X[X['anomaly'] == 1]
outliers = X[X['anomaly'] == -1]

print(f"Number of inliers: {len(inliers)}")
print(f"Number of outliers: {len(outliers)}")

display(outliers.head())
```

8. Train a SARIMA model on the monthly airline passengers dataset.
Forecast the next 12 months and visualize the results.


```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
import warnings

warnings.filterwarnings("ignore") # Suppress warnings related to convergence

# For the AirPassengers dataset, common SARIMA orders are often (p,d,q)(P,D,Q,s)
# A commonly used model for AirPassengers is (1,1,1)(1,1,0,12) or similar.
# Let's try with (1,1,1)(1,1,1,12) as a robust starting point.
# (p,d,q) = non-seasonal order
# (P,D,Q,s) = seasonal order (s=12 for monthly data)

order = (1, 1, 1)        # Non-seasonal p, d, q
seasonal_order = (1, 1, 1, 12) # Seasonal P, D, Q, s

print(f"Training SARIMA model with order={order} and seasonal_order={seasonal_order}")

# Fit the SARIMA model
sarima_model = SARIMAX(airline['Passengers'],
                       order=order,
                       seasonal_order=seasonal_order,
                       enforce_stationarity=False,
                       enforce_invertibility=False)

sarima_results = sarima_model.fit(disp=False)

print(sarima_results.summary())
```

Now, let's forecast the next 12 months and visualize the results.

```python
# Get the last date in the original dataset
last_date = airline.index[-1]

# Create a date range for the next 12 months
forecast_start_date = last_date + pd.DateOffset(months=1)
forecast_end_date = last_date + pd.DateOffset(months=12)
forecast_index = pd.date_range(start=forecast_start_date, end=forecast_end_date, freq='MS') # 'MS' for month start

# Generate forecast
forecast = sarima_results.get_forecast(steps=12)
forecast_df = forecast.conf_int()
forecast_df['mean'] = forecast.predicted_mean
forecast_df.index = forecast_index

# Plot the original series, fitted values, and the forecast
plt.figure(figsize=(14, 7))
plt.plot(airline['Passengers'], label='Original Data')
plt.plot(sarima_results.fittedvalues, color='green', label='Fitted Values')
plt.plot(forecast_df['mean'], color='red', label='Forecast')
plt.fill_between(forecast_df.index,
                 forecast_df.iloc[:, 0],
                 forecast_df.iloc[:, 1],
                 color='pink', alpha=0.3, label='Confidence Interval')

plt.title('Air Passengers SARIMA Forecast')
plt.xlabel('Date')
plt.ylabel('Number of Passengers')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.neighbors import LocalOutlierFactor
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure df_taxi is available from previous steps
# If not, you might need to re-run the cell where df_taxi is loaded or dummy data is created.
if 'df_taxi' not in locals():
    print("df_taxi not found. Please run the cell to load NYC Taxi Fare Data (or create dummy data) first.")
    # For demonstration, creating a dummy DataFrame if df_taxi is not defined
    import numpy as np
    np.random.seed(42)
    data = {
        'pickup_longitude': np.random.uniform(-74.05, -73.95, 1000),
        'pickup_latitude': np.random.uniform(40.7, 40.8, 1000),
        'dropoff_longitude': np.random.uniform(-74.05, -73.95, 1000),
        'dropoff_latitude': np.random.uniform(40.7, 40.8, 1000),
        'fare_amount': np.random.normal(12, 5, 1000),
        'passenger_count': np.random.randint(1, 7, 1000)
    }
    df_taxi = pd.DataFrame(data)
    df_taxi.loc[100, 'fare_amount'] = 150 # High fare
    df_taxi.loc[101, 'passenger_count'] = 10 # High passenger count
    df_taxi.loc[102, ['pickup_longitude', 'pickup_latitude']] = [-70.0, 35.0] # Outlier location
    df_taxi.loc[103, 'fare_amount'] = 0.5 # Very low fare

# Select features for anomaly detection
features = ['fare_amount', 'passenger_count']
X_lof = df_taxi[features].copy()

# Handle potential missing values (if any)
X_lof = X_lof.dropna()

# Initialize and train the Local Outlier Factor model
# n_neighbors: Number of neighbors to consider for the local density estimation.
#              Commonly 20 is a good starting point.
# contamination: The proportion of outliers in the data set.
lof = LocalOutlierFactor(n_neighbors=20, contamination='auto')

# Fit the model and predict anomaly scores (negative_outlier_factor_)
# Note: LOF's predict method returns -1 for outliers and 1 for inliers directly
# or use fit_predict which returns the anomaly labels (-1 or 1).
# The 'decision_function' is equivalent to -lof.negative_outlier_factor_.

y_pred = lof.fit_predict(X_lof)
X_lof['anomaly'] = y_pred

# Get the raw anomaly scores (higher is more normal, lower is more anomalous)
X_lof['lof_score'] = lof.negative_outlier_factor_

# Separate inliers and outliers for plotting
inliers_lof = X_lof[X_lof['anomaly'] == 1]
outliers_lof = X_lof[X_lof['anomaly'] == -1]

print(f"Number of inliers detected by LOF: {len(inliers_lof)}")
print(f"Number of outliers detected by LOF: {len(outliers_lof)}")

display(outliers_lof.head())
```

Finally, we will visualize the anomalies detected by LOF on a 2D scatter plot.

```python
# Visualize the anomalies
plt.figure(figsize=(10, 6))
sns.scatterplot(x='fare_amount', y='passenger_count', data=inliers_lof, color='blue', label='Normal points', alpha=0.6)
sns.scatterplot(x='fare_amount', y='passenger_count', data=outliers_lof, color='red', label='Anomalies', s=100, marker='X')

plt.title('Anomaly Detection using Local Outlier Factor (Fare Amount vs. Passenger Count)')
plt.xlabel('Fare Amount')
plt.ylabel('Passenger Count')
plt.legend()
plt.grid(True)
plt.show()
```

### Real-time Data Science Workflow for Power Grid Monitoring

As a data scientist for a power grid monitoring company, my goal is to forecast energy demand and detect abnormal spikes or drops in real-time consumption data collected every 15 minutes. The dataset includes features like timestamp, region, weather conditions, and energy usage. Here's how I would approach the real-time data science workflow:

---

#### 1. Anomaly Detection in Streaming Data

For real-time anomaly detection in streaming data, the choice among Isolation Forest, LOF, and DBSCAN depends on the characteristics of anomalies expected and computational constraints. Given 15-minute interval data, we need models that can process data efficiently.

*   **Initial Approach (Batch-like for new data windows):**
    *   **Isolation Forest:** This would be my primary choice due to its efficiency with high-dimensional data and its ability to work well without explicitly defining data clusters. It's particularly good at isolating anomalies quickly. In a streaming context, we would likely train an Isolation Forest model on recent historical data (e.g., the last few weeks/months) and then use this pre-trained model to score incoming batches of 15-minute data. When a new batch arrives, new data points are passed through the existing trees, and anomaly scores are computed. Anomalies are flagged if their scores fall below a pre-defined threshold.
    *   **DBSCAN:** While excellent for detecting arbitrarily shaped clusters and outliers in spatial data, DBSCAN can be computationally more intensive for large datasets and requires careful tuning of its `eps` (neighborhood distance) and `min_samples` parameters, which might need re-tuning as data patterns evolve. It's less ideal for high-speed, continuous streaming where model retraining is frequent.
    *   **Local Outlier Factor (LOF):** LOF is good for varied density datasets where the anomaly score depends on the local neighborhood. However, like DBSCAN, it can be computationally expensive for high-volume streaming data as it involves calculating distances to neighbors for each new point, which can be slow if the reference dataset is large.

*   **Real-time Adaptation:**
    *   For truly real-time, continuous processing, a hybrid approach might be necessary. A lightweight, efficient model (like a statistical process control method or a simple thresholding rule) could serve as a first pass, and flagged instances could then be fed to a more sophisticated model like Isolation Forest for deeper analysis, potentially on a slightly delayed micro-batch.
    *   The Isolation Forest model would be periodically re-trained (e.g., daily or weekly) on a rolling window of the most recent historical data to adapt to changing consumption patterns and grid behavior.

---

#### 2. Time Series Model for Short-Term Forecasting

For short-term forecasting (e.g., next 1-4 hours) of energy demand with 15-minute interval data, considering features like timestamp, region, and weather conditions:

*   **SARIMAX (Seasonal AutoRegressive Integrated Moving Average with eXogenous regressors) Model:** This would be the most suitable choice.
    *   **Seasonality (SARIMA component):** Energy demand typically exhibits strong daily, weekly, and possibly even monthly seasonality (e.g., higher demand during peak hours, weekdays vs. weekends). SARIMA explicitly accounts for these seasonal patterns.
    *   **Exogenous Regressors (X component):** Weather conditions (temperature, humidity, cloud cover) are critical drivers of energy demand. SARIMAX allows incorporating these external factors, significantly improving forecast accuracy. Other exogenous variables could include holidays, special events, or regional economic indicators.
    *   **Differencing (Integrated component):** Energy usage time series are often non-stationary, exhibiting trends or changing variance. The 'Integrated' component handles this by differencing the data to make it stationary before applying AR and MA components.

*   **Why not AR/MA/ARIMA?**
    *   **AR/MA:** These are too simplistic and do not handle seasonality or exogenous variables, which are crucial for accurate energy demand forecasting.
    *   **ARIMA:** While better than AR/MA, it lacks the ability to explicitly model seasonality and incorporate external factors, which would lead to less accurate forecasts for this type of data.

*   **Implementation:** Multiple SARIMAX models could be trained, one for each region, to capture regional-specific demand patterns and weather impacts.

---

#### 3. Validation and Monitoring Performance Over Time

Robust validation and monitoring are crucial for real-time systems:

*   **Initial Validation (Offline):**
    *   **Backtesting:** Evaluate model performance on historical data using a rolling-origin cross-validation setup. Train the model on data up to a certain point and forecast the next period, then roll forward. This simulates real-time deployment.
    *   **Metrics:** Use appropriate time series forecasting metrics such as Mean Absolute Error (MAE), Root Mean Squared Error (RMSE), Mean Absolute Percentage Error (MAPE), and Symmetric Mean Absolute Percentage Error (sMAPE). For anomaly detection, metrics like Precision, Recall, F1-score, and AUC-ROC (if labeled anomaly data is available) would be used.

*   **Real-time Monitoring (Online):**
    *   **Drift Detection:** Monitor for data drift (changes in input data distribution, e.g., weather patterns, new grid infrastructure) and concept drift (changes in the relationship between inputs and outputs, e.g., demand response programs altering consumption habits). Statistical tests (e.g., ADWIN) or comparison of feature distributions can be used.
    *   **Performance Monitoring:** Continuously track forecast accuracy (e.g., daily/weekly MAE, RMSE) and anomaly detection effectiveness (e.g., false positive rate, false negative rate) on new, incoming data. Compare these against predefined thresholds.
    *   **Model Retraining Strategy:** Implement an automated retraining pipeline. If performance metrics degrade or significant data/concept drift is detected, the model should be automatically retrained on the most recent, relevant data. A fixed schedule (e.g., weekly retraining) combined with event-triggered retraining (on drift detection) is a robust strategy.
    *   **Alerting:** Set up alerts for model performance degradation, high anomaly rates, or system failures to ensure prompt human intervention.

---

#### 4. Business Decisions and Operations Impact

This real-time data science solution would provide significant value to business decisions and operations:

*   **Optimized Grid Management:**
    *   **Load Balancing:** Accurate short-term demand forecasts enable grid operators to better anticipate and manage electricity generation and distribution, ensuring stable supply and preventing overload.
    *   **Resource Allocation:** Optimize the dispatch of power plants and renewable energy sources, minimizing operational costs and maximizing efficiency.

*   **Proactive Anomaly Response:**
    *   **Fault Detection:** Early detection of abnormal drops in energy usage could indicate equipment failure or outages, allowing for rapid response and restoration of service.
    *   **Theft Detection:** Unexplained spikes or unusual patterns might signal energy theft or unauthorized usage, leading to investigation and prevention.
    *   **Maintenance Scheduling:** Persistent anomalies in specific regions or equipment could suggest impending failures, informing predictive maintenance schedules and reducing costly unplanned downtime.
